```python
# Functions + Type Hints
from typing import List

def mean(values: List[float]) -> float:
    """
    Calculates the arithmetic mean of a list of numbers.

    Args:
        values: A list of floats or integers.

    Returns:
        The average of the values as a float.

    Raises:
        ValueError: If the input list is empty.
    """
    if not values:
        raise ValueError("Cannot calculate the mean of an empty list.")
    return sum(values) / len(values)
print(mean([10.5, 20.0, 30.5]))
# print(mean([]))  

```

```python
from typing import List
def dot_product(a: List[float], b: List[float]) -> float:
   """
    Computes the dot product of two numerical vectors.

    The dot product is the sum of the products of the corresponding entries 
    of the two sequences of numbers (a[0]*b[0] + a[1]*b[1] ...).

    Args:
        a: A list of floats representing the first vector.
        b: A list of floats representing the second vector.

    Returns:
        The resulting dot product as a float.

    Raises:
        ValueError: If the input lists have different lengths.
    """
    if len(a) != len(b):
        raise ValueError("Vectors must have the same length")
    result = 0.0

    for i in range(len(a)):
        result += a[i] * b[i]
    return result
print(dot_product([1, 2, 3],[4, 5, 6]))
```
