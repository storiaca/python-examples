```python
# Benchmark It, loop
import numpy as np
import time

a = np.array([1,2,3])
b = np.array([4,5,6])

size = 1_000_000

a = np.random.rand(size)
b = np.random.rand(size)

# NumPy
start = time.time()
np.dot(a, b)
print("NumPy:", time.time() - start)

# Python loop
a_list = a.tolist()
b_list = b.tolist()

start = time.time()
sum(x*y for x,y in zip(a_list, b_list))
print("Python:", time.time() - start)

# NumPy: 0.005675077438354492
# Python: 0.063812255859375
```
